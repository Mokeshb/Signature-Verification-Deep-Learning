```python
import os
import numpy as np
from PIL import Image

def process_images_from_directory(directory_path):
    images = []
    labels = []
    print(f"Loading images from {directory_path}...")

    # Traverse each subdirectory in the main directory
    for subdir in os.listdir(directory_path):
        subdir_path = os.path.join(directory_path, subdir)
        if os.path.isdir(subdir_path):  # Ensure it is a directory
            print(f"Processing subdirectory: {subdir}")
            for filename in os.listdir(subdir_path):
                if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                    image_path = os.path.join(subdir_path, filename)
                    try:
                        # Load and process the image
                        image = Image.open(image_path).convert('L')  # Convert to grayscale
                        image = image.resize((256, 256))  # Resize to 256x256
                        image = np.array(image) / 255.0  # Normalize pixel values
                        images.append(image)

                        # Assign label based on filename
                        label = 1 if 'forge' in filename.lower() else 0
                        labels.append(label)
                    except Exception as e:
                        print(f"Error processing file {image_path}: {e}")

    print(f"Processed {len(images)} images.")
    return np.array(images), np.array(labels)

def load_dataset(train_dir, test_dir):
    print("Loading training dataset...")
    train_images, train_labels = process_images_from_directory(train_dir)

    print("Loading testing dataset...")
    test_images, test_labels = process_images_from_directory(test_dir)

    return (train_images, train_labels), (test_images, test_labels)

# Example usage with dataset paths
train_dir = "/content/archive/Train"
test_dir = "/content/archive/Test"

(train_images, train_labels), (test_images, test_labels) = load_dataset(train_dir, test_dir)

# Output dataset shapes for verification
print(f"Training images shape: {train_images.shape}")
print(f"Training labels shape: {train_labels.shape}")
print(f"Testing images shape: {test_images.shape}")
print(f"Testing labels shape: {test_labels.shape}")
```

```python
import os
import subprocess
from PIL import Image
import numpy as np

def process_images_with_metadata(directory_path):
    images = []
    labels = []
    print(f"Loading from {directory_path}...")

    # Traverse each subdirectory in the main directory
    for subdir in os.listdir(directory_path):
        subdir_path = os.path.join(directory_path, subdir)
        if os.path.isdir(subdir_path):  # Ensure it is a directory
            print(f"Entering subdirectory: {subdir}")
            for filename in os.listdir(subdir_path):
                if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                    image_path = os.path.join(subdir_path, filename)

                    # Extract metadata using the 'file' command
                    try:
                        result = subprocess.run(['file', image_path], capture_output=True, text=True)
                        print(f"Metadata for {filename}: {result.stdout.strip()}")
                    except subprocess.CalledProcessError as e:
                        print(f"Failed to get metadata for {filename}: {e}")
                        continue  # Skip files if the command fails

                    # Process the image
                    try:
                        image = Image.open(image_path).convert('L')  # Convert to grayscale
                        image = image.resize((256, 256))  # Resize to 256x256
                        image = np.array(image) / 255.0  # Normalize pixel values
                        images.append(image)

                        # Assign label based on filename
                        label = 1 if 'forge' in filename.lower() else 0
                        labels.append(label)
                    except Exception as e:
                        print(f"Error processing image {filename}: {e}")

    print(f"Processed {len(images)} images.")
    return np.array(images), np.array(labels)

# Define the paths to the extracted directories
train_dir = "/content/archive/Train"
test_dir = "/content/archive/Test"

# Process directories
print("Processing training dataset...")
train_images, train_labels = process_images_with_metadata(train_dir)

print("\nProcessing testing dataset...")
test_images, test_labels = process_images_with_metadata(test_dir)

# Output dataset shapes for verification
print(f"Training images shape: {train_images.shape}")
print(f"Training labels shape: {train_labels.shape}")
print(f"Testing images shape: {test_images.shape}")
print(f"Testing labels shape: {test_labels.shape}")
```

```python
import os
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import matplotlib.pyplot as plt

# Define the model
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 1)),
    MaxPooling2D(pool_size=(2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')  # Binary classification
])

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='rmsprop', metrics=['accuracy'])
model.summary()

# Train the model
history = model.fit(
    train_images, train_labels,
    epochs=10,  # Number of epochs
    batch_size=32,  # Mini-batch size
    validation_split=0.2  # Use 20% of training data for validation
)

# Evaluate the model on the test dataset
test_loss, test_accuracy = model.evaluate(test_images, test_labels)
print(f"Test Accuracy: {test_accuracy * 100:.2f}%")

# Visualize training progress
# Accuracy plot
plt.figure(figsize=(12, 5))
plt.plot(history.history['accuracy'], label='Training Accuracy', marker='o')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy', marker='o')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()

# Loss plot
plt.figure(figsize=(12, 5))
plt.plot(history.history['loss'], label='Training Loss', marker='o')
plt.plot(history.history['val_loss'], label='Validation Loss', marker='o')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Function to plot the training and validation accuracy and loss
def plot_training_history(history):
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = range(1, len(acc) + 1)

    # Plot accuracy
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, 'bo-', label='Training Accuracy')  # Blue with dots and lines
    plt.plot(epochs, val_acc, 'ro-', label='Validation Accuracy')  # Red with dots and lines
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.grid(True)

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, 'bo-', label='Training Loss')
    plt.plot(epochs, val_loss, 'ro-', label='Validation Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.grid(True)

    plt.tight_layout()
    plt.show()

# Call the function with your history object
plot_training_history(history)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to display random test images with predictions
def display_random_predictions(images, labels, predictions, num_samples=10):
    indices = np.random.choice(len(images), num_samples, replace=False)  # Randomly select indices
    plt.figure(figsize=(10, 10))
    for i, idx in enumerate(indices):
        plt.subplot(num_samples // 2, 2, i + 1)
        plt.imshow(images[idx].reshape(256, 256), cmap='gray')
        actual_label = "Original" if labels[idx] == 0 else "Forged"
        predicted_label = "Original" if predictions[idx][0] < 0.5 else "Forged"
        plt.title(f'Actual: {actual_label}\nPredicted: {predicted_label}')
        plt.axis('off')
    plt.tight_layout()
    plt.show()

# Get predictions and display random samples
predictions = model.predict(test_images)
display_random_predictions(test_images, test_labels, predictions)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def display_predictions(images, labels, predictions, num_images=20):
    """
    Displays a grid of randomly selected test images with their actual and predicted labels.

    Args:
        images (numpy.ndarray): Test images.
        labels (numpy.ndarray): Actual labels (0 = Original, 1 = Forged).
        predictions (numpy.ndarray): Predicted probabilities for the images.
        num_images (int): Number of images to display. Default is 20.
    """
    plt.figure(figsize=(15, 15))  # Set the figure size
    n = min(num_images, len(images))  # Ensure 'num_images' doesn't exceed the dataset size

    # Randomly select 'n' unique indices
    selected_indices = np.random.choice(len(images), size=n, replace=False)

    # Display each randomly selected image
    for idx, i in enumerate(selected_indices, 1):
        plt.subplot(5, 4, idx)  # 5 rows and 4 columns
        plt.imshow(images[i].reshape(256, 256), cmap='gray')  # Display the grayscale image
        actual_label = 'Original' if labels[i] == 0 else 'Forged'
        predicted_label = 'Original' if predictions[i] < 0.5 else 'Forged'
        plt.title(f'Actual: {actual_label}\nPredicted: {predicted_label}')
        plt.axis('off')

    plt.tight_layout()  # Adjust layout for better spacing
    plt.show()

# Evaluate the model on the test set
test_scores = model.evaluate(test_images, test_labels)
print(f'Test Accuracy: {test_scores[1]:.2f}')

# Predict on the test set
predictions = model.predict(test_images)

# Display predictions
display_predictions(test_images, test_labels, predictions)
```

```python
from sklearn.metrics import classification_report

pred_labels = (predictions >= 0.5).astype(int).flatten()
print(classification_report(test_labels, pred_labels, target_names=['Original', 'Forged']))
```
